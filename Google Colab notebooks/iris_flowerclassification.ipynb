```python
import numpy as np
import matplotlib
from sklearn import datasets

iris = datasets.load_iris()
print(iris)

X = iris.data
print(X)


```

```python
Y = iris.target
print(Y)
```

```python
print(X.shape)
```

```python
print(Y.shape)
```

```python
#the first 10 rows of data 
X[:10]
```

```python
Y[:10]
```

```python
#firs coloumn of data 
a = X[:,[0,2]]
import matplotlib.pyplot as plt
plt.plot(X[:,2],X[:,0],'or')

# plt.scatter(a[:,1],a[:,0])
plt.scatter(X[:,2],X[:,0])

```

```python
# specify labels
# separate the data with the same label
#label 0 data are given in red
#label 1 data are given in green
#label 2 data are given in magenta

plt.plot(X[Y[:]==0,2],X[Y[:]==0,0],'or')
plt.plot(X[Y[:]==1,2],X[Y[:]==1,0],'og')
plt.plot(X[Y[:]==2,2],X[Y[:]==2,0],'om')
plt.title('Iris data')
plt.xlabel('Petal.Length')
plt.ylabel('Sepal.Length')
plt.legend(['Setosa','Versicolor','Virginica'])
plt.show()
```

```python
#data splitting: 2/3 for training, 1/3 for testing
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,Y,test_size=0.33)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Decision tree model
from sklearn import tree
classifier = tree.DecisionTreeClassifier()

```

```python
#train the model 
clf = classifier.fit(X_train,y_train)#only training data
```

```python
import pydotplus
from IPython.display import Image
dot_data = tree.export_graphviz(clf, out_file=None,feature_names=iris.feature_names ,class_names=iris.target_names ,filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
graph.write_png("iris.png")

from IPython.display import Image
#Image("iris.png")
Image(filename = "iris.png", width = 600)
```

```python
#prediction using test data
predictions = classifier.predict(X_test)
print(predictions)

print(y_test)
print((predictions!=y_test)/50)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,predictions))
```

```python
#prediction using entire dataset
clf = classifier.fit(X,Y)
import pydotplus
from IPython.display import Image
dot_data = tree.export_graphviz(clf, out_file=None,feature_names=iris.feature_names ,class_names=iris.target_names ,filled=True, rounded=True,special_characters=True)
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
graph.write_png("iris.png")

from IPython.display import Image
#Image("iris.png")
Image(filename = "iris.png", width = 600)

# in order to get an average learning model with its parameters have to run numerous time e.g. 10^4
```
